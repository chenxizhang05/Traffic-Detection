Assumptions:
Only two independent random samples with at least ordinally scaled characteristics must be available.
The variables do not have to fulfill a distribution curve

Null hypothesis:
In the population, the sum of the rankings in the two group does not differ

Alternaltive hypothesis:
In the population, the sum of the ranking differs in the two groups

```python
file_path = '/content/data_final (1).xlsx'
data1 = pd.read_excel(file_path, sheet_name='20230130')
data2 = pd.read_excel(file_path, sheet_name='20230208')
data3 = pd.read_excel(file_path, sheet_name='20230202')
data4 = pd.read_excel(file_path, sheet_name='20230203')
data5 = pd.read_excel(file_path, sheet_name='20230214')
data6 = pd.read_excel(file_path, sheet_name='20230316')
data7 = pd.read_excel(file_path, sheet_name='20230228')
data8 = pd.read_excel(file_path, sheet_name='20230302')
```

```python
class1 = data1['model-vehicles'].values
class2 = data2['model-vehicles'].values
class3 = data3['model-vehicles'].values/24
class4 = data4['model-vehicles'].values/24
class5 = data5['model-vehicles'].values/32
class6 = data6['model-vehicles'].values/32
class7 = data7['model-vehicles'].values/9
class8 = data8['model-vehicles'].values/6
```

Mann-Whitney U test (non-parametric)

```python
# Perform Mann-Whitney U test
statistic_1, p_value_1 = mannwhitneyu(class1, class2, alternative='two-sided')
statistic_2, p_value_2 = mannwhitneyu(class3, class4, alternative='two-sided')
statistic_3, p_value_3 = mannwhitneyu(class5, class6, alternative='two-sided')
statistic_4, p_value_4 = mannwhitneyu(class7, class8, alternative='two-sided')
# Output the results
print("If the p-value is below your threshold of significance, then you can reject the null hypothesis,\nbut this does not necessarily mean that your alternative hypothesis is true.")
print("Mann-Whitney U statistic:", statistic_1)
print("Two-sided p-value:", p_value_1)
print("Mann-Whitney U statistic:", statistic_2)
print("Two-sided p-value:", p_value_2)
print("Mann-Whitney U statistic:", statistic_3)
print("Two-sided p-value:", p_value_3)
print("Mann-Whitney U statistic:", statistic_4)
print("Two-sided p-value:", p_value_4)
```

T-test (Parametric)

```python
# t_statistic, p_value = ttest_ind(class1, class2)

# Welch's t-test (independent two-sample t-test with unequal variances)
t_statistic_1, p_value_1 = ttest_ind(class1, class2, equal_var=False)
t_statistic_2, p_value_2 = ttest_ind(class3, class4, equal_var=False)
t_statistic_3, p_value_3 = ttest_ind(class5, class6, equal_var=False)
t_statistic_4, p_value_4 = ttest_ind(class7, class8, equal_var=False)

# Print the results
# print("If the p-value is below your threshold of significance, then you can reject the null hypothesis,\nbut this does not necessarily mean that your alternative hypothesis is true.")

print("T-statistic:", t_statistic_1)
print("Two-sided p-value:", p_value_1)
print("T-statistic:", t_statistic_2)
print("Two-sided p-value:", p_value_2)
print("T-statistic:", t_statistic_3)
print("Two-sided p-value:", p_value_3)
print("T-statistic:", t_statistic_4)
print("Two-sided p-value:", p_value_4)
```
