```python
base_path = '/content/gdrive/MyDrive/chenxi_streamline/20230203'
# insert video_files in the directory
video_files = ['GX020030.MP4']
frameRate = 60  # it will capture image every 0.5 seconds

for video_file in video_files:
# video_file = 'GX010030.MP4'

  path = str(os.path.join(base_path, video_file[:-4]))
  # Create a directory for each video file
  os.makedirs(path, exist_ok=True)
  images_dir = str(os.path.join(base_path, video_file))
  vidcap = cv2.VideoCapture(images_dir)
  def getFrame(sec):
    vidcap.set(cv2.CAP_PROP_POS_MSEC,sec*1000)
    hasFrames,image = vidcap.read()
    if hasFrames:
      image_path = str(os.path.join(path, f"Image{count}.jpg"))
      # print(image_path)
      cv2.imwrite(image_path, image)
      # cv2.imwrite('/content/gdrive/MyDrive/chenxi_streamline/20230203/GX010030/Image' + str(count)+'.jpg', image) # save frame as JPG file
    return hasFrames
  sec = 0
  frameRate = 60 #//it will capture image in each 0.5 second
  count=1
  success = getFrame(sec)
  while success:
    count = count + 1
    sec = sec + frameRate
    sec = round(sec, 2)
    success = getFrame(sec)
  print(f"Processed video '{video_file}'. Total frames: {count - 1}")

  IMAGES_FOR_TEST = {}
  for i in range(1, count):
      image_name = f'Image {i}'
      image_path = str(os.path.join(path, f'Image{i}.jpg'))
      IMAGES_FOR_TEST[image_name] = image_path

  image_list = [Image.open(IMAGES_FOR_TEST[f'Image {i}']) for i in range(1, count)]

  def crop_image(index, x, y, width, height):
      area = (x, y, width, height)
      cropped_img = image_list[index].crop(area)
      # cropped_img.show()
      crop_path = str(os.path.join(path, f"Cropped_Image{index+1}.jpg"))
      cropped_img.save(crop_path)

  for i in range(count-1):
    crop_image(i, 400, 600, 1200, 1000)
```
