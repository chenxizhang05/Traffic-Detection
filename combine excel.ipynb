```python
import pandas as pd
import glob

# Get a list of all Excel files in the directory
excel_files = glob.glob("/content/gdrive/MyDrive/A*STAR Traffic Project/20230214/*.xlsx")

# Initialize an empty dataframe to store the combined data
combined_df = pd.DataFrame()

# Iterate over the Excel files and append their data to the combined dataframe
for file in excel_files:
    df = pd.read_excel(file)
    combined_df = combined_df.append(df, ignore_index=True)

# Write the combined dataframe to a new Excel file
combined_excel_path = "/content/gdrive/MyDrive/A*STAR Traffic Project/20230214/20230214_final.xlsx"
combined_df.to_excel(combined_excel_path, index=False)
```
